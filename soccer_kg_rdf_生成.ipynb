### 命名空间

### 生成关系实体

```python
with open("soccerQuintuples.csv", "r", encoding="utf-8") as f:
    facts = csv.reader(f)
    name = "null"
    node = None
    count = 0
    for fact in facts:
        count += 1
#         print(fact)
        s = fact[0].strip()
        p = fact[1]
        o = fact[2].strip()
#         print((s, p, o))
        if p == "效力于":
            ts, te = fact[3], fact[4]
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = next(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            g.add((tail, RDF.type, SOCCER.Team))
            
            relation = BNode()
            g.add((head, SOCCER.Predicate, relation))
            g.add((relation, SOCCER.Object, tail))
            g.add((relation, SOCCER.含义, Literal("效力")))
            g.add((relation, RDF.type, SOCCER.PlayFor))
            
            interval = BNode()
            g.add((relation, SOCCER.HappendIn, interval))
            g.add((interval, RDF.type, SOCCER.Interval))
            g.add((interval, SOCCER.start, Literal(ts)))
            g.add((interval, SOCCER.end, Literal(te)))    
        if p == "荣誉":
            ts, te = fact[3], fact[4]
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = next(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            g.add((tail, RDF.type, SOCCER.Honor))
            
            relation = BNode()
            g.add((head, SOCCER.Predicate, relation))
            g.add((relation, SOCCER.Object, tail))
            g.add((relation, SOCCER.含义, Literal("荣誉")))
            g.add((relation, RDF.type, SOCCER.HasHonor))
            
            interval = BNode()
            g.add((relation, SOCCER.HappendIn, interval))
            g.add((interval, RDF.type, SOCCER.Interval))
            g.add((interval, SOCCER.start, Literal(ts)))
            g.add((interval, SOCCER.end, Literal(te)))
        if p == "伤病":
            ts, te = fact[3], fact[4]
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = next(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            g.add((tail, RDF.type, SOCCER.Injury))
            
            relation = BNode()
            g.add((head, SOCCER.Predicate, relation))
            g.add((relation, SOCCER.Object, tail))
            g.add((relation, SOCCER.含义, Literal("伤病")))
            g.add((relation, RDF.type, SOCCER.HasInjury))
            
            interval = BNode()
            g.add((relation, SOCCER.HappendIn, interval))
            g.add((interval, RDF.type, SOCCER.Interval))
            g.add((interval, SOCCER.start, Literal(ts)))
            g.add((interval, SOCCER.end, Literal(te)))  
            
        if p == "教练":
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = list(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            if tail:
                tail = tail[0]
            else:
                tail = URIRef(baseUri + o.strip().replace(" ", "・"))
                g.add((tail, SOCCER.名字, Literal(o.strip())))
            g.add((head, SOCCER.HasCoach, tail))
            g.add((tail, RDF.type, SOCCER.Coach))
            
        if p == "前锋":
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = list(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            if tail:
                tail = tail[0]
            else:
                tail = URIRef(baseUri + o.strip().replace(" ", "・"))
                g.add((tail, SOCCER.名字, Literal(o.strip())))
            g.add((head, SOCCER.前锋, tail))
            g.add((tail, RDF.type, SOCCER.Forward))
        
        if p == "中场":
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = list(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            if tail:
                tail = tail[0]
            else:
                tail = URIRef(baseUri + o.strip().replace(" ", "・"))
                g.add((tail, SOCCER.名字, Literal(o.strip())))
            g.add((head, SOCCER.中场, tail))
            g.add((tail, RDF.type, SOCCER.Midfield))
            
        if p == "后卫":
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = list(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            if tail:
                tail = tail[0]
            else:
                tail = URIRef(baseUri + o.strip().replace(" ", "・"))
                g.add((tail, SOCCER.名字, Literal(o.strip())))
            g.add((head, SOCCER.后卫, tail))
            g.add((tail, RDF.type, SOCCER.Back))
            
        if p == "门将":
            head = next(g.subjects(predicate=SOCCER.名字, object=Literal(s)))
            tail = list(g.subjects(predicate=SOCCER.名字, object=Literal(o)))
            if tail:
                tail = tail[0]
            else:
                tail = URIRef(baseUri + o.strip().replace(" ", "・"))
                g.add((tail, SOCCER.名字, Literal(o.strip())))
            g.add((head, SOCCER.门将, tail))
            g.add((tail, RDF.type, SOCCER.Goalkeeper))
```
