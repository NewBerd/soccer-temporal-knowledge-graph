```python
#!/usr/bin/env python3
# coding: utf-8
# File: question_classifier.py
# Author: lhy<lhy_in_blcu@126.com,https://huangyong.github.io>
# Date: 18-10-4

import os
import ahocorasick

class QuestionClassifier:
    def __init__(self):
#         cur_dir = '/'.join(os.path.abspath(__file__).split('/')[:-1])
        cur_dir = "./"
        #　特征词路径
        self.players_path = os.path.join(cur_dir, "playersName.txt")
        self.teams_path = os.path.join(cur_dir, "teamsName.txt")
        self.honors_path = os.path.join(cur_dir, "honorsName.txt")
        self.injury_name_path = os.path.join(cur_dir, "injuryName.txt")
        self.team_path = os.path.join(cur_dir, "qwds/team.txt")
        self.country_path = os.path.join(cur_dir, "qwds/country.txt")
        self.age_path = os.path.join(cur_dir, "qwds/age.txt")
        self.height_path = os.path.join(cur_dir, "qwds/height.txt")
        self.weight_path = os.path.join(cur_dir, "qwds/weight.txt")
        self.position_path = os.path.join(cur_dir, "qwds/position.txt")
        self.number_path = os.path.join(cur_dir, "qwds/number.txt")
        self.birthday_path = os.path.join(cur_dir, "qwds/birthday.txt")
        self.customfoot_path = os.path.join(cur_dir, "qwds/customfoot.txt")
        self.capability_path = os.path.join(cur_dir, "qwds/capability.txt")
        self.honor_path = os.path.join(cur_dir, "qwds/honor.txt")
        self.injury_path = os.path.join(cur_dir, "qwds/injury.txt")
        self.rongna_path = os.path.join(cur_dir, "qwds/rongna.txt")
        self.establish_path = os.path.join(cur_dir, "qwds/establish.txt")
        self.home_path = os.path.join(cur_dir, "qwds/home.txt")
        self.city_path = os.path.join(cur_dir, "qwds/city.txt")
        self.phone_path = os.path.join(cur_dir, "qwds/phone.txt")
        self.email_path = os.path.join(cur_dir, "qwds/email.txt")
        self.location_path = os.path.join(cur_dir, "qwds/location.txt")
        self.coach_path = os.path.join(cur_dir, "qwds/coach.txt")
        self.forward_path = os.path.join(cur_dir, "qwds/forward.txt")
        self.middle_path = os.path.join(cur_dir, "qwds/middle.txt")
        self.backward_path = os.path.join(cur_dir, "qwds/backward.txt")
        self.goalkeeper_path = os.path.join(cur_dir, "qwds/goalkeeper.txt")
        self.time_path = os.path.join(cur_dir, "qwds/time.txt")
        

        
        # 加载特征词/实体
        self.players_wds = [wd.rstrip() for wd in open(self.players_path, "r", encoding="utf-8") if wd.rstrip()]
        self.teams_wds = [wd.rstrip() for wd in open(self.teams_path, "r", encoding="utf-8") if wd.rstrip()]
        self.honors_wds = [wd.rstrip() for wd in open(self.honors_path, "r", encoding="utf-8") if wd.rstrip()]
        self.injury_wds = [wd.rstrip() for wd in open(self.injury_name_path, "r", encoding="utf-8") if wd.rstrip()]
        self.region_words = list(set(self.players_wds + self.honors_wds + self.teams_wds + self.injury_wds))
#         self.deny_words = [i.strip() for i in open(self.deny_path, "r", encoding="utf-8") if i.strip()]

        # 构造领域actree
        self.region_tree = self.build_actree(list(self.region_words))
        
        # 构建词典：实体名词：类别
        self.wdtype_dict = self.build_wdtype_dict()
        
        # 问句疑问词
        self.team_qwds = [wd.rstrip() for wd in open(self.team_path, "r", encoding="utf-8") if wd.rstrip()]
        self.country_qwds = [wd.rstrip() for wd in open(self.country_path, "r", encoding="utf-8") if wd.rstrip()]
        self.age_qwds = [wd.rstrip() for wd in open(self.age_path, "r", encoding="utf-8") if wd.rstrip()]
        self.height_qwds = [wd.rstrip() for wd in open(self.height_path, "r", encoding="utf-8") if wd.rstrip()]
        self.weight_qwds = [wd.rstrip() for wd in open(self.weight_path, "r", encoding="utf-8") if wd.rstrip()]
        self.positon_qwds = [wd.rstrip() for wd in open(self.position_path, "r", encoding="utf-8") if wd.rstrip()]
        self.number_qwds = [wd.rstrip() for wd in open(self.number_path, "r", encoding="utf-8") if wd.rstrip()]
        self.birthday_qwds = [wd.rstrip() for wd in open(self.birthday_path, "r", encoding="utf-8") if wd.rstrip()]
        self.customfoot_qwds = [wd.rstrip() for wd in open(self.customfoot_path, "r", encoding="utf-8") if wd.rstrip()]
        self.capability_qwds = [wd.rstrip() for wd in open(self.capability_path, "r", encoding="utf-8") if wd.rstrip()]
        self.honor_qwds = [wd.rstrip() for wd in open(self.honor_path, "r", encoding="utf-8") if wd.rstrip()]
        self.injury_qwds = [wd.rstrip() for wd in open(self.injury_path, "r", encoding="utf-8") if wd.rstrip()]
        self.rongna_qwds = [wd.rstrip() for wd in open(self.rongna_path, "r", encoding="utf-8") if wd.rstrip()]
        self.establish_qwds = [wd.rstrip() for wd in open(self.establish_path, "r", encoding="utf-8") if wd.rstrip()]
        self.home_qwds = [wd.rstrip() for wd in open(self.home_path, "r", encoding="utf-8") if wd.rstrip()]
        self.city_qwds = [wd.rstrip() for wd in open(self.city_path, "r", encoding="utf-8") if wd.rstrip()]
        self.phone_qwds = [wd.rstrip() for wd in open(self.phone_path, "r", encoding="utf-8") if wd.rstrip()]
        self.email_qwds = [wd.rstrip() for wd in open(self.email_path, "r", encoding="utf-8") if wd.rstrip()]
        self.location_qwds = [wd.rstrip() for wd in open(self.location_path, "r", encoding="utf-8") if wd.rstrip()]
        self.coach_qwds = [wd.rstrip() for wd in open(self.coach_path, "r", encoding="utf-8") if wd.rstrip()]
        self.forward_qwds = [wd.rstrip() for wd in open(self.forward_path, "r", encoding="utf-8") if wd.rstrip()]
        self.middle_qwds = [wd.rstrip() for wd in open(self.middle_path, "r", encoding="utf-8") if wd.rstrip()]
        self.backward_qwds = [wd.rstrip() for wd in open(self.backward_path, "r", encoding="utf-8") if wd.rstrip()]
        self.goalkeeper_qwds = [wd.rstrip() for wd in open(self.goalkeeper_path, "r", encoding="utf-8") if wd.rstrip()]
        self.time_qwds = [wd.rstrip() for wd in open(self.time_path, "r", encoding="utf-8") if wd.rstrip()]

        self.club_qwds = ["球队", "俱乐部", "队", "队伍"]
        self.people_qwds = ["运动员", "球员", "谁", "哪个人", "哪些人"]
        #print('model init finished ......')
        return

    '''分类主函数'''
    def classify(self, question):
        data = {}
        entity_dict = self.check_entity(question)
        if not entity_dict:
            return {}
        data['args'] = entity_dict
        #收集问句当中所涉及到的实体类型
        types = []
        for type_ in entity_dict.values():
            types += type_
        question_type = 'others'

        question_types = []

        # 俱乐部/球队
        if self.isAppear(self.team_qwds, question) and ('players' in types):
            question_type = 'players_club'
            question_types.append(question_type)
#         国籍/国家
        if self.isAppear(self.country_qwds, question) and ('players' in types):
            question_type = 'players_country'
            question_types.append(question_type)
        if self.isAppear(self.country_qwds, question) and ("teams" in types):
            question_type = "teams_country"
            question_types.append(question_type)           
#         身高
        if self.isAppear(self.height_qwds, question) and ("players" in types):
            question_type = "players_height"
            question_types.append(question_type)
#         体重
        if self.isAppear(self.weight_qwds, question) and ("players" in types):
            question_type = "players_weight"
            question_types.append(question_type)
#         年龄
        if self.isAppear(self.age_qwds, question) and ("players" in types):
            question_type = "players_age"
            question_types.append(question_type)
#        位置（中锋/前锋）
        if self.isAppear(self.positon_qwds, question) and ("players" in types):
            question_type = "players_position"
            question_types.append(question_type)
# 号码
        if self.isAppear(self.number_qwds, question) and ("players" in types):
            question_type = "players_number"
            question_types.append(question_type)        
# 生日
        if self.isAppear(self.birthday_qwds, question) and ("players" in types):
            question_type = "players_birthday"
            question_types.append(question_type)
# 惯用脚
        if self.isAppear(self.customfoot_qwds, question) and ("players" in types):
            question_type = "players_customfoot"
            question_types.append(question_type)
# 能力
        if self.isAppear(self.capability_qwds, question) and ("players" in types):
            question_type = "players_capability"
            question_types.append(question_type)      
# 荣誉
        if self.isAppear(self.honor_qwds, question) and ("players" in types):
            question_type = "players_honor"
            question_types.append(question_type) 
        if self.isAppear(self.honor_qwds, question) and ("teams" in types):
            question_type = "teams_honor"
            question_types.append(question_type) 
# 伤病
        if self.isAppear(self.injury_qwds, question) and ("players" in types):
            question_type = "players_injury"
            question_types.append(question_type)
# 容纳
        if self.isAppear(self.rongna_qwds, question) and ("teams" in types):
            question_type = "teams_rongna"
            question_types.append(question_type)
# 成立
        if self.isAppear(self.establish_qwds, question) and ("teams" in types):
            question_type = "teams_establish"
            question_types.append(question_type)
# 主场
        if self.isAppear(self.home_qwds, question) and ("teams" in types):
            question_type = "teams_home"
            question_types.append(question_type)
# 城市
        if self.isAppear(self.city_qwds, question) and ("teams" in types):
            question_type = "teams_city"
            question_types.append(question_type)
# 电话号码
        if self.isAppear(self.phone_qwds, question) and ("teams" in types):
            question_type = "teams_phone"
            question_types.append(question_type)
# 邮箱
        if self.isAppear(self.email_qwds, question) and ("teams" in types):
            question_type = "teams_email"
            question_types.append(question_type)
# 地址
        if self.isAppear(self.location_qwds, question) and ("teams" in types):
            question_type = "teams_location"
            question_types.append(question_type)
# 教练
        if self.isAppear(self.coach_qwds, question) and ("teams" in types):
            question_type = "teams_coach"
            question_types.append(question_type)
# 前场
        if self.isAppear(self.forward_qwds, question) and ("teams" in types):
            question_type = "teams_forward"
            question_types.append(question_type)
# 中场
        if self.isAppear(self.middle_qwds, question) and ("teams" in types):
            question_type = "teams_middle"
            question_types.append(question_type)
# 后卫
        if self.isAppear(self.backward_qwds, question) and ("teams" in types):
            question_type = "teams_backward"
            question_types.append(question_type)
# 门将
        if self.isAppear(self.goalkeeper_qwds, question) and ("teams" in types):
            question_type = "teams_goalkeeper"
            question_types.append(question_type)
#         某个奖项有哪些人获得
        if self.isAppear(self.people_qwds, question) and "honors" in types:
            question_type = "honors_players"
            question_types.append(question_type)
            
            #         某个奖项有哪些教练获得
        if self.isAppear(self.coach_qwds, question) and "honors" in types:
            question_type = "honors_coaches"
            question_types.append(question_type)
            
            #         某个奖项有哪些前锋获得
        if self.isAppear(self.forward_qwds, question) and "honors" in types:
            question_type = "honors_forwards"
            question_types.append(question_type)
            
            #         某个奖项有哪些中场获得
        if self.isAppear(self.middle_qwds, question) and "honors" in types:
            question_type = "honors_middles"
            question_types.append(question_type)
            
            #         某个奖项有哪些后卫获得
        if self.isAppear(self.backward_qwds, question) and "honors" in types:
            question_type = "honors_backwards"
            question_types.append(question_type)
            
            #         某个奖项有哪些门将获得
        if self.isAppear(self.goalkeeper_qwds, question) and "honors" in types:
            question_type = "honors_goalkeepers"
            question_types.append(question_type)
            
        # 某个奖项有哪些球队获得
        if self.isAppear(self.club_qwds, question) and "honors" in types:
            question_type = "honors_teams"
            question_types.append(question_type)
            
#         某个伤病有哪些人获得
        if self.isAppear(self.people_qwds, question) and "injury" in types:
            question_type = "injury_players"
            question_types.append(question_type)
            
            #         某个伤病有哪些教练获得
        if self.isAppear(self.coach_qwds, question) and "injury" in types:
            question_type = "injury_coaches"
            question_types.append(question_type)
            
            #         某个伤病有哪些前锋获得
        if self.isAppear(self.forward_qwds, question) and "injury" in types:
            question_type = "injury_forwards"
            question_types.append(question_type)
            
            #         某个伤病有哪些中场获得
        if self.isAppear(self.middle_qwds, question) and "injury" in types:
            question_type = "injury_middles"
            question_types.append(question_type)
            
            #         某个伤病有哪些后卫获得
        if self.isAppear(self.backward_qwds, question) and "injury" in types:
            question_type = "injury_backwards"
            question_types.append(question_type)
            
            #         某个伤病有哪些门将获得
        if self.isAppear(self.goalkeeper_qwds, question) and "injury" in types:
            question_type = "injury_goalkeepers"
            question_types.append(question_type)            

#         运动员效力球队的时间
        if self.isAppear(self.time_qwds, question) and "players" in types and "teams" in types:
            question_type = "players_teams_time"
            question_types.append(question_type)
            
            #         运动员获得荣誉的时间
        if self.isAppear(self.time_qwds, question) and "players" in types and "honors" in types:
            question_type = "players_honors_time"
            question_types.append(question_type)
 
        # 某种伤病的时间
        if self.isAppear(self.time_qwds, question) and "players" in types and "injury" in types:
            question_type = "players_injury_time"
            question_types.append(question_type)
    
#         球队荣誉时间
        if self.isAppear(self.time_qwds, question) and "teams" in types and "honors" in types:
            question_type = "teams_honors_time"
            question_types.append(question_type)
        
        # 若没有查到相关的外部查询信息，那么则将该球员的描述信息返回
        if question_types == [] and 'players' in types:
            question_types = ["players_birthday","players_age", "players_height", "players_weight", 
                              "players_country", "players_club", "players_position", 
                             "players_number", "players_customfoot", "players_honor"]

        # 若没有查到相关的外部查询信息，那么则将该球队的描述信息返回
        if question_types == [] and 'teams' in types:
            question_types = ["teams_establish", "teams_city", "teams_home", "teams_phone", "teams_email",
                             "teams_coach", "teams_forward", "teams_middle", "teams_backward", 
                              "teams_goalkeeper", "teams_honor"]
        
        

        # 将多个分类结果进行合并处理，组装成一个字典
        data['question_types'] = question_types

        return data

    '''构造实体词对应的类型映射'''
    def build_wdtype_dict(self):
        wd_dict = {}
        for wd in self.players_wds:
            wd_dict[wd] = ["players"]
        for wd in self.teams_wds:
            wd_dict[wd] = ["teams"]
        for wd in self.honors_wds:
            wd_dict[wd] = ["honors"]
        for wd in self.injury_wds:
            wd_dict[wd] = ["injury"]
#         print(wd_dict)
        return wd_dict

    '''构造actree，加速过滤'''
    def build_actree(self, wordlist):
        actree = ahocorasick.Automaton()
        for index, word in enumerate(wordlist):
            actree.add_word(word, (index, word))
        actree.make_automaton()
        return actree

    '''实体识别/特征词识别'''
    def check_entity(self, question):
        region_wds = []
        for i in self.region_tree.iter(question):
            wd = i[1][1]
            region_wds.append(wd)
        stop_wds = []
        for wd1 in region_wds:
            for wd2 in region_wds:
                if wd1 in wd2 and wd1 != wd2:
                    stop_wds.append(wd1)
        final_wds = [i for i in region_wds if i not in stop_wds]
        final_dict = {i:self.wdtype_dict.get(i) for i in final_wds}

        return final_dict

    '''检查特征词是否出现在sent问句中'''
    def isAppear(self, wds, sent):
        for wd in wds:
            if wd in sent:
                return True
        return False

```

```python
from SPARQLWrapper import SPARQLWrapper, JSON
from collections import OrderedDict

class AnswerSearcher:
#     http://172.22.105.72:3030/soccer_temporal_kg
    def __init__(self, endpoint_url='http://172.22.105.72:3030/soccer_temporal_kg/query'):
        self.sparql_conn = SPARQLWrapper(endpoint_url)
        self.prefix = '''PREFIX SOCCER: <http://www.soccer_temporal_kg.com#>
                            PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
                            PREFIX owl: <http://www.w3.org/2002/07/owl#>
                            PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
                            PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
                            PREFIX type: <http://info.deepcarbon.net/schema/type#>
                        '''
        
    def search_main(self, sql_result):
        final_answers = []
        for item in sql_result:
            question_type = item['question_type']
            for sql_ in item['sql']:
                sql, subject = sql_[0], sql_[1]
                query_result = self.get_sparql_result(self.prefix + sql)
                print("query_result:", query_result)
                answers = {"subject":subject, "ans":query_result}
                nl_ans = self.answer_pretty(question_type, answers)
                if nl_ans != "":
                    final_answers.append(nl_ans)
        return final_answers
            

    #     套用模板使答案美化      
    def answer_pretty(self, question_type, answers):
        nl_ans = ""
        subject = answers.get("subject")
        query = answers.get("ans") #query是双层列表
        if not query:
            return nl_ans

        #         关于运动员的普通问题
        if question_type.startswith("players_") and not question_type.endswith("time"):
            if question_type == "players_height":
                nl_ans = "{}的身高是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_club":
                nl_ans = "{}的俱乐部是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_country":
                nl_ans = "{}的国籍是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_weight":
                nl_ans = "{}的体重是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_age":
                nl_ans = "{}的年龄是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_position":
                nl_ans = "{}的场上位置是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_number":
                nl_ans = "{}的球衣号码是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_birthday":
                nl_ans = "{}的生日是{}".format(subject, "、".join(query[0]))

            elif question_type == "players_customfoot":
                nl_ans = "{}的惯用脚是{}".format(subject, "、".join(query[0]))
                
            elif question_type == "players_capability":
                nl_ans = "{}的综合能力评分为{}".format(subject, "、".join(query[0]))
                
            elif question_type == "players_honor":
                nl_ans = "{}所获得的荣誉有{}".format(subject, "、".join(query[0]))
                
            elif question_type == "players_injury":
                nl_ans = "{}曾受过的伤病：{}".format(subject, "、".join(query[0]))
        
#         关于球队的普通问题
        elif question_type.startswith("teams_") and not question_type.endswith("time"):
            if question_type == "teams_country":
                nl_ans = "{}是{}的俱乐部".format(subject, "、".join(query[0]))

            elif question_type == "teams_rongna":
                nl_ans = "{}的主场可以容纳观众{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_establish":
                nl_ans = "{}的成立时间是{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_home":
                nl_ans = "{}的主场是{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_city":
                nl_ans = "{}所在城市为{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_phone":
                nl_ans = "{}的联系电话是{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_email":
                nl_ans = "{}的电子邮箱是{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_location":
                nl_ans = "{}在{}".format(subject, "、".join(query[0]))

            elif question_type == "teams_coach":
                nl_ans = "{}的教练：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "teams_forward":
                nl_ans = "{}的前锋：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "teams_middle":
                nl_ans = "{}的中场：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "teams_backward":
                nl_ans = "{}的后卫：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "teams_goalkeeper":
                nl_ans = "{}的门将：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "teams_honor":
                nl_ans = "{}的荣誉：{}".format(subject, "、".join(query[0]))
        elif question_type.startswith("honors_"):
            if question_type == "honors_teams":
                nl_ans = "获得荣誉“{}”的俱乐部：{}".format(subject, "、".join(query[0]))

            elif question_type == "honors_players":
                nl_ans = "获得荣誉“{}”的运动员：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "honors_coaches":
                nl_ans = "获得荣誉“{}”的教练员：{}".format(subject, "、".join(query[0]))
        
            elif question_type == "honors_forwards":
                nl_ans = "获得荣誉“{}”的前锋：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "honors_middles":
                nl_ans = "获得荣誉“{}”的中场球员：{}".format(subject, "、".join(query[0]))
        
            elif question_type == "honors_backwards":
                nl_ans = "获得荣誉“{}”的后卫：{}".format(subject, "、".join(query[0]))
                
            elif question_type == "honors_goalkeepers":
                nl_ans = "获得荣誉“{}”的守门员：{}".format(subject, "、".join(query[0]))
        
        elif question_type.startswith("injury_"):
            if question_type == "injury_players":
                nl_ans = "受过伤病{}的运动员有：{}".format(subject, "、".join(query[0]))

            elif question_type == "injury_coaches":
                nl_ans = "受过伤病{}的教练员有：{}".format(subject, "、".join(query[0]))

            elif question_type == "injury_forwards":
                nl_ans = "受过伤病{}的前锋有：{}".format(subject, "、".join(query[0]))
            
            elif question_type == "injury_middles":
                nl_ans = "受过伤病{}的中场球员有：{}".format(subject, "、".join(query[0]))
            
            elif question_type == "injury_backwards":
                nl_ans = "受过伤病{}的后卫有：{}".format(subject, "、".join(query[0]))
            
            elif question_type == "injury_goalkeepers":
                nl_ans = "受过伤病{}的守门员有：{}".format(subject, "、".join(query[0]))
        
        elif question_type.endswith("time"):
            if question_type == "players_honors_time":
                nl_ans = "{}获得荣誉{}的时间为:".format(subject[0], subject[1]) + "、".join(["{}年-{}年".format(s, e) for s, e in zip(query[0], query[1])])
            
            elif question_type == "players_teams_time":
                nl_ans = "{}在俱乐部{}效力的时间为:".format(subject[0], subject[1]) + "、".join(["{}年-{}年".format(s, e) for s, e in zip(query[0], query[1])])
            
            elif question_type == "players_injury_time":
                nl_ans = "{}{}的时间为:".format(subject[0], subject[1]) + "、".join(["{}年-{}年".format(s, e) for s, e in zip(query[0], query[1])])
            
            elif question_type == "teams_honors_time":
                nl_ans = "{}俱乐部获得荣誉{}的时间为:".format(subject[0], subject[1]) + "、".join(["{}年-{}年".format(s, e) for s, e in zip(query[0], query[1])])
        
        return nl_ans
    
    
    #     取查询结果
    # 
    # { "head": {
    #     "vars": [ "subject" , "name" ]
    #   } ,
    #   "results": {
    #     "bindings": [
    #       { 
    #         "subject": { "type": "uri" , "value": "http://www.soccer_temporal_kg.com#戴维-洛佩斯" } ,
    #         "name": { "type": "literal" , "value": "戴维-洛佩斯" }
    #       } ,
    #       { 
    #         "subject": { "type": "uri" , "value": "http://www.soccer_temporal_kg.com#皮克" } ,
    #         "name": { "type": "literal" , "value": "皮克" }
    #       }
    #     ]
    #   }
    # }
#     return：[[value1,..],[val1,..],..]
# 
    def get_sparql_result(self, query):
        self.sparql_conn.setQuery(query)
        self.sparql_conn.setReturnFormat(JSON)
        query_result = self.sparql_conn.query().convert()
        
        results = dict()
        col = query_result["head"]["vars"]
        
        for e in col:
            results[e] = []
#         将列名和列值对应起来 {列名：[列值1,], }   
        for item in query_result["results"]["bindings"]:
            for k, v in item.items():
                results[k].append(v["value"])

        #         返回不带列名的结果，以便于分析
        result = []
        for e in col:
            result.append(results.get(e))
        return result

    
        
```
