```python
import csv
```

```python
with open("soccerQuintuples.csv", "r", encoding="utf-8") as f:
    facts = csv.reader(f)
    entity = []
    team = []
    for fact in facts:
        p = fact[1]
        o = fact[2]
        if p in ("效力于", "荣誉", "伤病"):
            entity.append(o)
    entity = set(entity)
    print(entity)
```
